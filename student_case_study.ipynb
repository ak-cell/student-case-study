```python
#importing Library
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
#Reading Data
data=pd.read_csv("C:/Users/aksha/Desktop/ai/datasets/student.csv")
print(data.shape)
data.head()
```

```python
# Collecting X and Y
X=data['Math'].values
Y=data['Reading'].values
```

```python
#Mean X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)
#Total number of values
m=len(X)
#Using the formula to calculate b1 and b2
numer=0
denom=0
for i in range (m):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
b1=numer/denom 
b0=mean_y-(b1*mean_x)
#print coefficients
print(b1,b0)

    
```

```python
#Plotting Values in the regression line
max_x=np.max(X)+100
min_x=np.min(X)-100
#Calculating Line Values x and y
x=np.linspace(min_x,max_x,1000)
y=b0+b1*x
#Plotting Line
plt.plot(x,y,c='Red',label='Regression Line')
#Plotting scatter plots
plt.scatter(X,Y,c='Green',label='Scatter Plot')
plt.xlabel('Math')
plt.ylabel('Reading')
plt.legend()
plt.show()
```

```python
#Calculating the Root Mean Square Error
rmse=0
for i in range (m):
    y_pred=b0+b1*X[i]
    rmse+=(Y[i]-y_pred)**2
rmse=np.sqrt(rmse/m)
print(rmse)
```

```python
ss_t=0   #Total sum of square
ss_r=0 #Total sum of square of residuals
for i in range (m):
    y_pred=b0+b1*X[i]
    ss_t+=(Y[i]-mean_y)**2
    ss_r+=(Y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
print(r2)
```

```python
#using Machine Learning 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#Cannot use Rank1 matrix in scikit learn
X=X.reshape((m,1))
#Creating Model
reg=LinearRegression()
#FittingTrainingData
reg=reg.fit(X,Y)
#Y Prediction
Y_pred=reg.predict(X)
# CalculATING RMSE and R2 Score
mse=mean_squared_error(Y,Y_pred)
rmse=np.sqrt(mse)
r2_score=reg.score(X,Y)
print(np.sqrt(mse))
print(r2_score)
```
